```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
import seaborn as sns 
```

```python
avg_mgw_df = pd.read_csv("../data/Trj2Shape_3n4m.artifact_purged.avg.MGW", 
                         header='infer', 
                         sep="\s+")
```

```python
avg_mgw_df
```

```python
print ("We can see that the first and last three ResId have many missing points", 
       "so we try to discard them while plotting")
```

```python
subset_avg_mgw_df = avg_mgw_df[3:17]
```

```python
subset_avg_mgw_df
```

```python
sns.set()
mgw_plt = sns.pointplot(data=subset_avg_mgw_df, x="ResId", y="Mean")
```

```python
x_coords = []
y_coords = []
for point_pair in mgw_plt.collections:
    for x, y in point_pair.get_offsets():
        x_coords.append(x)
        y_coords.append(y)
# Credits: https://stackoverflow.com/questions/43159528/error-bars-with-seaborn-and-stripplot
```

```python
mgw_plt.errorbar(x_coords, y_coords, yerr=subset_avg_mgw_df["Std"], ecolor="red", fmt=' ', zorder=-1)
```

```python
mgw_plt.figure
```

```python
mgw_plt.figure.savefig("avg_MGW_profile.png")
```
